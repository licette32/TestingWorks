### üåü Desaf√≠o Extra en Python

üß† ¬øCu√°l es la condici√≥n de existencia de un tri√°ngulo?

Un conjunto de tres lados forma un tri√°ngulo si y solo si:

ùëé + ùëè > ùëê ,

ùëé + ùëê > ùëè ,

ùëè + ùëê > ùëé


Es decir:

‚úî La suma de dos lados debe ser mayor al tercer lado.

‚úî Esto debe cumplirse para las tres combinaciones.

<br>

üõ†Ô∏è Nuevo dise√±o de la funci√≥n

Ahora nuestra funci√≥n `determinarTriangulo` debe:

1. Primero verificar si los lados forman un tri√°ngulo real.

2. Si NO forman un tri√°ngulo, retornar -1 (o lo que vos quieras definir).

3. Si forman un tri√°ngulo, clasificar (equil√°tero, is√≥sceles, escaleno) como antes.

```python
# --- Funci√≥n determinarTriangulo mejorada con validaci√≥n de tri√°ngulo ---
def determinarTriangulo(a, b, c):
    """
    Determina el tipo de tri√°ngulo dado sus lados a, b y c.
    Retornar:
        -1 --> No forman un tri√°ngulo v√°lido
         1 --> Is√≥sceles
         2 --> Escaleno
         3 --> Equil√°tero
    """
    # Validar si es un tri√°ngulo real
    if (a + b <= c) or (a + c <= b) or (b + c <= a):
        return -1  # No es un tri√°ngulo v√°lido

    # Clasificar el tipo de tri√°ngulo
    if a == b == c:
        return 3  # Equil√°tero
    elif a == b or a == c or b == c:
        return 1  # Is√≥sceles
    else:
        return 2  # Escaleno

# --- Funci√≥n de test mejorada ---
def test():
    """
    Funci√≥n de prueba para determinarTriangulo.
    Imprime el resultado de cada prueba y un resumen final.
    """
    # Casos de prueba
    casos_prueba = [
        (5, 5, 5),   # Equil√°tero
        (6, 2, 6),   # Is√≥sceles
        (6, 7, 8),   # Escaleno
        (1, 1, 3),   # No forma tri√°ngulo
        (2, 2, 3),   # Is√≥sceles
        (10, 1, 1)   # No forma tri√°ngulo
    ]

    resultados_esperados = [3, 1, 2, -1, 1, -1]

    todo_ok = True

    for i in range(len(casos_prueba)):
        a, b, c = casos_prueba[i]
        resultado = determinarTriangulo(a, b, c)
        if resultado == resultados_esperados[i]:
            print(f"‚úÖ Caso {i+1}: ({a}, {b}, {c}) -> Correcto")
        else:
            print(f"‚ùå Caso {i+1}: ({a}, {b}, {c}) -> Error: esperado {resultados_esperados[i]}, obtenido {resultado}")
            todo_ok = False

    # Resumen final
    if todo_ok:
        print("\nüéâ Todos los casos de prueba pasaron correctamente.")
        return 1
    else:
        print("\n‚ö†Ô∏è Hubo errores en algunas pruebas.")
        return 0

# --- Ejecutamos ---
test()
```

Qu√© cambia con esto?

Antes, cualquier combinaci√≥n se clasificaba como tri√°ngulo, en cambio ahora se clasifican combinaciones v√°lidas. Ahora, con la validaci√≥n detectamos si los lados no forman un tri√°ngulo (antes no detect√°bamos errores de lados).

```python
%%R
# Funci√≥n para determinar el tipo de tri√°ngulo
determinar_triangulo <- function(a, b, c) {
  if (a == b && b == c) {
    return(3)  # Equil√°tero
  }
  else if (a == b || a == c || b == c) {
    return(1)  # Is√≥sceles
  }
  else {
    return(2)  # Escaleno
  }
}

# Funci√≥n de prueba para verificar la implementaci√≥n
test_triangulo <- function() {
  # Definimos nuestros casos de prueba
  casos_prueba <- list(
    list(a = 2, b = 2, c = 2, esperado = 3),               # Equil√°tero
    list(a = 3, b = 3, c = 4, esperado = 1),               # Is√≥sceles
    list(a = 3, b = 4, c = 5, esperado = 2),               # Escaleno
    list(a = 5, b = 5, c = 5, esperado = 3),               # Equil√°tero
    list(a = 4, b = 4, c = 6, esperado = 1),               # Is√≥sceles
    list(a = 7, b = 8, c = 9, esperado = 2)                # Escaleno
  )

  # Probamos cada uno de los casos
  for (caso in casos_prueba) {
    resultado <- determinar_triangulo(caso$a, caso$b, caso$c)
    if (resultado != caso$esperado) {
      cat(sprintf("‚ùå Error con (%d, %d, %d): esperado %d, obtenido %d\n",
                 caso$a, caso$b, caso$c, caso$esperado, resultado))
      return(FALSE)
    }
  }

  cat("‚úÖ Todos los tests pasaron correctamente\n")
  return(TRUE)
}

# Ejecutamos los tests
test_triangulo()
```

```python
%%R
# Funci√≥n con validaci√≥n
determinarTriangulo <- function(a, b, c) {
  # Retorna:
  #  -1 --> No forman un tri√°ngulo v√°lido
  #   1 --> Is√≥sceles
  #   2 --> Escaleno
  #   3 --> Equil√°tero

  # Validar si los lados forman un tri√°ngulo real
  if (a + b <= c || a + c <= b || b + c <= a) {
    return(-1) # No es un tri√°ngulo v√°lido
  }

  # Clasificar el tipo de tri√°ngulo
  if (a == b && b == c) {
    return(3) # Equil√°tero
  } else if (a == b || a == c || b == c) {
    return(1) # Is√≥sceles
  } else {
    return(2) # Escaleno
  }
}

# Funci√≥n test
test <- function() {
  # Funci√≥n de prueba para determinarTriangulo

  casos_prueba <- list(
    c(5, 9, 5),   # Equil√°tero
    c(6, 2, 6),   # Is√≥sceles
    c(6, 7, 8),   # Escaleno
    c(1, 1, 6),   # No forman tri√°ngulo
    c(2, 2, 3),   # Is√≥sceles
    c(10, 1, 1)   # No forman tri√°ngulo
  )

  resultados_esperados <- c(3, 1, 2, -1, 1, -1)

  todo_ok <- TRUE

  for (i in 1:length(casos_prueba)) {
    lados <- casos_prueba[[i]]
    resultado <- determinarTriangulo(lados[1], lados[2], lados[3])
    if (resultado == resultados_esperados[i]) {
      cat(sprintf("‚úÖ Caso %d: (%d, %d, %d) -> Correcto\n", i, lados[1], lados[2], lados[3]))
    } else {
      cat(sprintf("‚ùå Caso %d: (%d, %d, %d) -> Error: esperado %d, obtenido %d\n",
                  i, lados[1], lados[2], lados[3], resultados_esperados[i], resultado))
      todo_ok <- FALSE
    }
  }

  if (todo_ok) {
    cat("\nüéâ Todos los casos de prueba pasaron correctamente.\n")
    return(1)
  } else {
    cat("\n‚ö†Ô∏è Hubo errores en algunas pruebas.\n")
    return(0)
  }
}

# Ejecutar pruebas
test()
```
